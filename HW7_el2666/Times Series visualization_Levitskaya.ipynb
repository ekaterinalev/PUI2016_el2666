## Historical NYC Crime Data: Times Series Visualization

### For this project, I wanted to take historical NYC crime data and plot time series visualization of fellonies conducted over time (from 2000 - 2015) in a category "Other Violations" (beyond the other category of "Harassment Offenses"). The original dataset is here: http://www.nyc.gov/html/nypd/downloads/excel/analysis_and_planning/violation_offenses_by_precinct_2000_2015.xls

### I started with a sample data from this dataset, where x = time, y = number of offenses.

```python
import matplotlib.pyplot as plt
import datetime
import numpy as np
```

```python
x = np.array([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015])
y = np.array([3614, 3593, 3510, 4187, 4299, 4126, 4741, 4501, 4371, 5094, 4291, 4582, 4086, 3383, 3125, 2213])
ymean = 3982.25
std = 688.9753715
error = std

ymin = y - std
ymax = y + std

plt.title('Violation Offenses Over Time (2000-2015)', fontsize=15, color="blue")
plt.xlabel('Years', fontsize=12)
plt.ylabel('Number of Offenses', fontsize=12)
plt.plot(x,y)
plt.fill_between(x, ymax, ymin, alpha=0.3)
plt.show()

```

### This plot allows us to see standard deviation and overall performance over the time, with spikes around year 2009 and then a dramatic decrease from year 2011 (4582 registered offenses) to 2015 (2213 registered offenses).
