### Cleaning data from "Unspecified" entries

```python
new_data = data[data['Community Board'] != '0 Unspecified']
```

```python
new_data1 = new_data[new_data['Community Board'] != 'Unspecified MANHATTAN']
```

```python
new_data2 = new_data1[new_data1['Community Board'] != 'Unspecified BROOKLYN']
```

```python
new_data3 = new_data2[new_data2['Community Board'] != 'Unspecified BRONX']
```

```python
new_data4 = new_data3[new_data3['Community Board'] != 'Unspecified QUEENS']
```

```python
new_data5 = new_data4[new_data4['Community Board'] != 'Unspecified STATEN ISLAND']
```

```python
ranking = new_data5.groupby('Community Board').count()
```

```python
ranking
```

```python
ranking.drop(['Descriptor'], axis=1, inplace=True)
```

```python
ranking.drop(['Agency Name'], axis=1, inplace=True)
```

```python
ranking.drop(['Borough'], axis=1, inplace=True)
```

### Creating ranking of Community Districtis by complaint type

```python
ranking['Rank'] = ranking['Complaint Type'].rank(ascending=1)
```

```python
ranking
```

### Sorting by Community Districts with highest number of complaints

```python
ranking_district = ranking.sort(['Rank'], ascending=False)
```

```python
ranking_district
```
